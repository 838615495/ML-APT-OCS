```python
import matplotlib.pyplot as plt

# 元素和对应的固定百分比数值
labels = ["Ti", "Zr", "Nb", "O", "N"]
values = [54.31, 37.02, 6.34, 1.43, 0.89]

# 配置柱子的颜色（边框）和 hatch 样式
colors = ['#FF8C42', '#3E8EDE', '#228B22', '#8B008B', '#FFD700']
hatches = ['////', '\\\\\\\\', 'xxxx', '....', '----']

# Matplotlib 样式设置
plt.rc('font', family='Times New Roman')
plt.rcParams.update({
    'font.weight': 'bold',
    'xtick.direction': 'in',
    'ytick.direction': 'in',
    'xtick.major.width': 3.5,
    'ytick.major.width': 3.5,
    'xtick.major.size': 5,
    'ytick.major.size': 5,
    'xtick.labelsize': 15,
    'ytick.labelsize': 15,
})

fig, ax = plt.subplots(figsize=(7, 3), dpi=600)

# 边框加粗
for spine in ax.spines.values():
    spine.set_linewidth(2)

# 绘制柱状图
x = range(len(labels))
width = 0.6
for i, (val, color, hatch) in enumerate(zip(values, colors, hatches)):
    ax.bar(x[i], val, width, color='white', edgecolor=color, hatch=hatch,
           linewidth=2, zorder=2)
    ax.text(x[i], val + 0.5, f'{val:.2f}%', ha='center', va='bottom',
            fontsize=12, fontweight='bold')

# 坐标轴设置
ax.set_xticks(x)
ax.set_xticklabels(labels, fontsize=20, fontweight='bold')
ax.set_ylabel('Atomic Fraction (%)', fontsize=15, fontweight='bold', labelpad=7)

# 刻度线
ax.tick_params(axis='x', which='major', length=5, width=2, direction='in')
ax.tick_params(axis='x', which='minor', length=3, width=1, direction='in')
ax.tick_params(axis='y', which='major', length=5, width=2, direction='in')

# y轴范围
ax.set_ylim(0, max(values) + 5)

plt.tight_layout()

plt.savefig("figure_b.pdf", bbox_inches='tight')  # ⭐ 保存为PDF（矢量）

plt.show()

import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

aspect_ratio = data_shape[:, 2]
oblateness = data_shape[:, 1]
atom_number_each_SRO = np.array(data_shape[:, -1], dtype=float)

# ✅ 
jitter_strength = 0.07  
np.random.seed(0)  # 可复现
aspect_ratio_jittered = aspect_ratio + np.random.uniform(-jitter_strength, jitter_strength, size=aspect_ratio.shape)
oblateness_jittered = oblateness + np.random.uniform(-jitter_strength, jitter_strength, size=oblateness.shape)

# 限制在合法范围内（0~1）
aspect_ratio_jittered = np.clip(aspect_ratio_jittered, 0, 1)
oblateness_jittered = np.clip(oblateness_jittered, 0, 1)

# ✅ 绘图
plt.rc('font', family='Times New Roman')
fig, ax = plt.subplots(dpi=600, figsize=(8, 6))

plt.rcParams['xtick.direction'] = 'in'
plt.rcParams['ytick.direction'] = 'in'
plt.xticks(fontproperties='Times New Roman', size=15, fontweight='bold')
plt.yticks(fontproperties='Times New Roman', size=15, fontweight='bold')
plt.tick_params(axis='x', width=1, length=5)
plt.tick_params(axis='y', width=1, length=7, which='major')
plt.tick_params(axis='y', width=0.5, length=4.5, which='minor')

ax.spines['bottom'].set_linewidth(2)
ax.spines['left'].set_linewidth(2)
ax.spines['top'].set_linewidth(2)
ax.spines['right'].set_linewidth(2)

# ✅ 使用蓝绿渐变配色：`winter`
sc = plt.scatter(oblateness_jittered, aspect_ratio_jittered, 
                 s=atom_number_each_SRO, c=atom_number_each_SRO, 
                 cmap='viridis', alpha=1,
                 vmin=0, vmax=160)

# 添加颜色条
cbar = plt.colorbar(sc)
cbar.set_label('Number of atoms', fontsize=18, fontweight='bold', labelpad=10)
cbar.ax.tick_params(labelsize=18)

# 添加标签
plt.xlabel('Oblateness', fontsize=18, fontweight='bold', labelpad=10)
plt.ylabel('Aspect ratio', fontsize=18, fontweight='bold', labelpad=10)
plt.tick_params(labelsize=18)

# 设置刻度
plt.xticks([0.0, 0.5, 1.0])
plt.yticks([0.0, 0.5, 1.0])

# 添加虚线框架
plt.vlines(0.5, 0, 1, linestyles="dashed")
plt.hlines(0.5, 0, 1, linestyles="dashed")

# 添加区域文本
plt.text(0.8, 0.6, 'Sphere', fontsize=18, fontweight='bold')
plt.text(1-0.8, 0.6, 'Disc', fontsize=18, fontweight='bold')
plt.text(1-0.8, 1-0.8, 'Lath', fontsize=18, fontweight='bold')
plt.text(0.8, 1-0.8, 'Rod', fontsize=18, fontweight='bold')
plt.tight_layout()

plt.savefig("figure_f_new.pdf", bbox_inches='tight')  # ⭐ 保存为PDF（矢量）

plt.show()
# 显示图形
plt.show()

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# 1. 数据
df = pd.DataFrame({
    "Interval": ["0-5", "5-10", "10-15", "15-20", "20-25", "25-30", "30-35", "35-40",
                 "40-45", "45-50", "50-55", "55-60", "60-65", "65-70", ">70"],
    "experiment": [381, 313, 132, 59, 41, 26, 23, 9, 13, 5, 4, 0, 3, 3, 7],
    "data_1":     [403, 334, 135, 60, 37, 27, 14, 13, 9, 3, 0, 1, 0, 1, 5],
    "data_2":     [393, 282, 118, 55, 30, 22, 7, 8, 7, 2, 1, 5, 1, 0, 1],
    "data_3":     [402, 304, 103, 63, 37, 22, 12, 6, 2, 0, 2, 3, 1, 1, 3]
})

# 2. 密度计算
N_total = 15625
df["experiment_density"] = df["experiment"] / N_total
df["data_1_density"] = df["data_1"] / N_total
df["data_2_density"] = df["data_2"] / N_total
df["data_3_density"] = df["data_3"] / N_total

df["random_mean_density"] = df[["data_1_density", "data_2_density", "data_3_density"]].mean(axis=1)

# 3. μ计算
chi_exp_total = ((df["experiment_density"] - df["random_mean_density"])**2 / df["random_mean_density"]).sum()
N_experiment = df["experiment_density"].sum()
mu_experiment = np.sqrt(chi_exp_total / N_experiment)

chi_rand_total = (
    (df[["data_1_density", "data_2_density", "data_3_density"]] -
     df["random_mean_density"].values.reshape(-1, 1))**2
    / df["random_mean_density"].values.reshape(-1, 1)
).sum().sum()

N_random = df[["data_1_density", "data_2_density", "data_3_density"]].sum().sum()
mu_random = np.sqrt(chi_rand_total / N_random)

# 4. 转换单位
df["experiment_density_m3"] = df["experiment_density"] * 1000
df["random_mean_density_m3"] = df["random_mean_density"] * 1000

# 5. 画柱状图（完全保持不变）
plt.rcParams['font.family'] = 'Times New Roman'
plt.rcParams['font.weight'] = 'bold'
plt.rcParams['xtick.direction'] = 'in'
plt.rcParams['ytick.direction'] = 'in'

fig, ax = plt.subplots(figsize=(5, 5), dpi=300)
bar_width = 0.4
x = np.arange(len(df))

ax.bar(x - bar_width/2, df["experiment_density_m3"],
       width=bar_width, label="Experiment Density", color="steelblue")
ax.bar(x + bar_width/2, df["random_mean_density_m3"],
       width=bar_width, label="Random Mean Density", color="orangered", alpha=0.7)

ax.spines['bottom'].set_linewidth(2)
ax.spines['left'].set_linewidth(2)
ax.spines['top'].set_linewidth(2)
ax.spines['right'].set_linewidth(2)

plt.tick_params(width=2, length=5)

ax.set_xlabel("Ti-Zr-rich domains", size=10, fontweight='bold', labelpad=7)
ax.set_ylabel("Number Density (10²⁴ m⁻³)", size=15, fontweight='bold', labelpad=7)
ax.set_xticks(x)
ax.set_xticklabels(df["Interval"], rotation=45, ha="right")

ax.legend(loc='upper right', bbox_to_anchor=(0.95, 0.55),
          framealpha=1, fontsize=12.5)

# 灰色区域
rectangle = plt.Rectangle(
    (3.5, 0),
    11,
    max(df["experiment_density_m3"] - 5),
    color="gray",
    alpha=0.2,
    zorder=0,
    edgecolor="k",
    linestyle="--"
)
ax.add_patch(rectangle)

# =========================
# ⭐ 在这里嵌入 μ 表格（核心）
# =========================

# 三条横线（用轴坐标！不会影响图）
ax.plot([0.28, 0.88], [0.92, 0.92], transform=ax.transAxes, color='black', lw=1)
ax.plot([0.28, 0.88], [0.84, 0.84], transform=ax.transAxes, color='black', lw=1)
ax.plot([0.28, 0.88], [0.76, 0.76], transform=ax.transAxes, color='black', lw=1)

# 表头
ax.text(0.40, 0.88, "Data", transform=ax.transAxes,
        ha='center', va='center', fontweight='bold')

ax.text(0.63, 0.88, "Random", transform=ax.transAxes,
        ha='center', va='center', fontweight='bold', color='orangered')

ax.text(0.82, 0.88, "Exp", transform=ax.transAxes,
        ha='center', va='center', fontweight='bold', color='steelblue')

# 数据行
ax.text(0.36, 0.80, "μ", transform=ax.transAxes,
        ha='center', va='center', fontweight='bold')

ax.text(0.63, 0.80, f"{mu_random:.3f}", transform=ax.transAxes,
        ha='center', va='center', fontweight='bold')

ax.text(0.82, 0.80, f"{mu_experiment:.3f}", transform=ax.transAxes,
        ha='center', va='center', fontweight='bold')

# =========================

plt.tight_layout()

plt.savefig("figure_e.pdf", bbox_inches='tight')  # ⭐ 保存为PDF（矢量）
plt.show()
```
