```python
import numpy as np
from scipy.spatial import cKDTree
from collections import defaultdict
from tqdm import tqdm

# 判断点是否在体素范围内
def in_voxel(pos, voxel_bounds):
    return all([
        voxel_bounds[0][0] <= pos[0] <= voxel_bounds[0][1],
        voxel_bounds[1][0] <= pos[1] <= voxel_bounds[1][1],
        voxel_bounds[2][0] <= pos[2] <= voxel_bounds[2][1]
    ])

# 计算体素内的SRO，使用邻居预缓存
def calculate_targeted_sro_pairs_cached(data, neighbors_idx, neighbors_dist, i,
                                        voxel_bounds, voxel_center, r_core,
                                        P_ge_m, structure_type_global,
                                        element_fractions, target_pairs):
    pos = data[i, :3]
    if np.linalg.norm(pos - voxel_center) > r_core:
        return None
    if not in_voxel(pos, voxel_bounds):
        return None

    center_type = int(data[i, 3])

    max_m_used = sum(p >= 0.01 for p in P_ge_m)
    valid_weights = P_ge_m[:max_m_used]

    neighbor_indices = neighbors_idx[i, 1:max_m_used+1]  # 第0是自己，跳过
    # neighbor_distances = neighbors_dist[i, 1:max_m_used+1]  # 可选用，当前未用

    neighbor_types = []
    applied_weights = []

    for rel_idx, j in enumerate(neighbor_indices):
        if j == i:
            continue
        neighbor_pos = data[j, :3]
        if not in_voxel(neighbor_pos, voxel_bounds):
            continue
        neighbor_types.append(int(data[j, 3]))
        applied_weights.append(valid_weights[rel_idx])

    if len(applied_weights) == 0:
        return {
            "xyz": tuple(pos),
            "atom_type": center_type,
            "sro_values": {f"{a}-{b}": np.nan for (a, b) in target_pairs},
            "structure_type": structure_type_global
        }

    weighted_counts = {}
    for t, w in zip(neighbor_types, applied_weights):
        pair_key = (center_type, t)
        if pair_key in target_pairs:
            weighted_counts[pair_key] = weighted_counts.get(pair_key, 0) + w

    total_weight = sum(applied_weights)
    sro_values = {}
    for (a, b) in target_pairs:
        if center_type != a:
            continue
        P_AB = weighted_counts.get((a, b), 0.0) / total_weight if total_weight > 0 else np.nan
        x_B = element_fractions.get(b, 1e-8)
        alpha_AB = P_AB / x_B if x_B > 0 else np.nan
        sro_values[f"{a}-{b}"] = alpha_AB

    return {
        "xyz": tuple(pos),
        "atom_type": center_type,
        "sro_values": sro_values,
        "structure_type": structure_type_global
    }

# 主计算函数：缓存邻居 + 体素筛选 + 进度条
def compute_sro_features_cached(data, data_sphere_points, P_ge_m,
                                r_core, structure_type_global,
                                element_fractions, target_pairs):
    tree = cKDTree(data[:, :3])
    max_m_used = sum(p >= 0.01 for p in P_ge_m)

    # 预查询所有点的邻居（包括自己，k=max_m_used+1）
    neighbors_dist, neighbors_idx = tree.query(data[:, :3], k=max_m_used+1)

    X_total = []
    for voxel_center in tqdm(data_sphere_points, desc="Processing voxels", unit="voxel"):
        voxel_bounds = (
            (voxel_center[0] - 0.5, voxel_center[0] + 0.5),
            (voxel_center[1] - 0.5, voxel_center[1] + 0.5),
            (voxel_center[2] - 0.5, voxel_center[2] + 0.5)
        )

        # 找出体素内点索引
        indices_in_voxel = [i for i, pos in enumerate(data[:, :3]) if in_voxel(pos, voxel_bounds)]

        sro_list = []
        for i in indices_in_voxel:
            sro_res = calculate_targeted_sro_pairs_cached(
                data, neighbors_idx, neighbors_dist, i,
                voxel_bounds, voxel_center, r_core,
                P_ge_m, structure_type_global,
                element_fractions, target_pairs)
            if sro_res:
                sro_list.append(sro_res)

        # 汇总计算
        pair_data = defaultdict(list)
        for atom in sro_list:
            sro_dict = atom.get("sro_values", {})
            for (a, b) in target_pairs:
                key = f"{a}-{b}"
                val = sro_dict.get(key, np.nan)
                if not np.isnan(val):
                    pair_data[key].append(val)

        feature_vector = []
        for (a, b) in target_pairs:
            key = f"{a}-{b}"
            vals = pair_data.get(key, [])
            mean_val = np.mean(vals) if len(vals) > 0 else 0.0
            feature_vector.append(mean_val)

        X_total.append(feature_vector)
    return np.array(X_total)
```

```python
import pandas as pd
import numpy as np

# 假设 df 已经存在，列为 ["x", "y", "z", 'mass', "type", "cluster_id"]

# 1️⃣ 按 cluster_id 统计每个团簇的原子数量
cluster_counts = df.groupby("cluster_id").size()

# 2️⃣ 先全部标记为 0（非 SRO）
df["cluster_type"] = 0
# 3️⃣ 筛选原子数量大于 25 的团簇
valid_clusters = cluster_counts[cluster_counts > 31].index
df_valid = df[df["cluster_id"].isin(valid_clusters)].copy()

# 4️⃣ 找出纯 Ti–Zr SRO 团簇（type 只包含 0,1,2）
pure_tizr_clusters = (
    df_valid.groupby("cluster_id")["type"]
    .apply(lambda types: set(types).issubset({0, 1, 2}))
)
pure_tizr_ids = pure_tizr_clusters[pure_tizr_clusters].index
df.loc[df["cluster_id"].isin(pure_tizr_ids), "cluster_type"] = 1

# 5️⃣ 对剩余的非纯 Ti–Zr 团簇，按 O/N 比例分类
mixed_clusters = valid_clusters.difference(pure_tizr_ids)

for cid in mixed_clusters:
    sub = df[df["cluster_id"] == cid]
    total_atoms = len(sub)
    o_count = (sub["type"] == 3).sum()
    n_count = (sub["type"] == 4).sum()
    
    o_ratio = o_count / total_atoms
    n_ratio = n_count / total_atoms
    
    if o_ratio > n_ratio :   # O 比 N 多，且差距大于 0.05
        ctype = 2
    elif n_ratio > o_ratio: # N 比 O 多
        ctype = 3
    else:                          # 差不多
        ctype = 4
    
    df.loc[df["cluster_id"] == cid, "cluster_type"] = ctype
# 最终结果
results = df.copy()
print(results.head())
```
